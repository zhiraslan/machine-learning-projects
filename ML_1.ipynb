```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("phones.csv")
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(10, 10)) #  
ax = plt.axes(projection="3d")
#    
ax.scatter(df["disk"], df["year"], df["price"], s=100)
plt.show()
```

###  

```python
from sklearn.linear_model import LinearRegression
```

```python
#        
X = df[["disk", "year"]]
y = df["price"]

#  
reg = LinearRegression().fit(X, y)
```

```python
#      
d1, d2 = list(), list()
for x in np.linspace(min(df["disk"]), max(df["disk"]), 100):
    for y in np.linspace(min(df["year"]), max(df["year"]), 100):
        d1.append(x)
        d2.append(y)
d1 = np.array(d1).reshape(-1, 1)
d2 = np.array(d2).reshape(-1, 1)
p = reg.predict(np.concatenate([d1, d2], axis=1))
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection = "3d")
ax.scatter(df["disk"], df["year"], df["price"], s=100)
ax.plot_trisurf(d1.ravel(), d2.ravel(), p.ravel(), alpha=0.2)

ax.elev=25
plt.show()
```
